```python
import os
from pathlib import Path

# Chemin vers le dossier QASMBench
target_path = Path("/content/drive/MyDrive/rl_quantum_circuit_routing/data/QASMBench")

if target_path.exists():
    print(f"✅ Le dossier existe : {target_path}")
    
    # 1. On compte tous les fichiers .qasm
    qasm_files = list(target_path.rglob("*.qasm"))
    print(f"📊 Total de fichiers .qasm trouvés : {len(qasm_files)}")
    
    # 2. On vérifie les sous-dossiers principaux (catégories)
    categories = [d.name for d in target_path.iterdir() if d.is_dir() and not d.name.startswith('.')]
    print(f"📂 Catégories détectées : {categories}")
    
    # 3. Petit aperçu du contenu pour être sûr
    print("\n🔍 Aperçu de quelques fichiers :")
    for file in qasm_files[:5]: # Affiche les 5 premiers
        print(f"  - {file.relative_to(target_path)}")
        
else:
    print("❌ Erreur : Le dossier est introuvable. Vérifie que le montage du Drive a bien fonctionné.")
```
